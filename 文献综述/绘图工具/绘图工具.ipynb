```python
import matplotlib.pyplot as plt
import numpy as np
def plot_paper_count(data):
    # Convert width from cm to inches
    width_in_inches = 12 / 2.54  # Change this to 15 for 15cm
    height_in_inches = 8 / 2.54  # You can adjust the height as needed

    plt.figure(figsize=(width_in_inches, height_in_inches))
    all_years = np.array([])
    all_counts = np.array([])
    for journal, (years, paper_counts) in data.items():
        plt.plot(years, paper_counts, label=journal)
        all_years = np.concatenate((all_years, years))
        all_counts = np.concatenate((all_counts, paper_counts))

    unique_years, indices = np.unique(all_years, return_inverse=True)
    total_counts = np.zeros_like(unique_years, dtype=int)
    np.add.at(total_counts, indices, all_counts)
    plt.plot(unique_years, total_counts, label="Total", linestyle="--")

    plt.xlabel("Year")
    plt.ylabel("Number of Papers")
    plt.title("Number of Papers per Year")
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
data = {
    "IEEE": (
        [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022],
        [3, 3, 2, 2, 1, 5, 6, 3, 4, 2],
    ),
    "ACM": (
        [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2023],
        [1, 1, 1, 3, 0, 2, 5, 0, 2, 1, 1],
    ),
    "Science Direct": ([2019, 2020, 2021, 2022, 2023], [1, 2, 2, 1, 4]),
    "Springer": (
        [2013, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024],
        [1, 2, 1, 4, 1, 2, 3, 9, 9, 6, 4],
    ),
}
plot_paper_count(data)
```

```python
total_papers = sum(sum(paper_counts) for _, paper_counts in data.values())
print(total_papers)
```
